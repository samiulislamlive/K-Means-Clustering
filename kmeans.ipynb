```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
iris=datasets.load_iris()
# print(iris.data) It will Give Only the Iris Data
# print(iris.target) It will Give Only the Target values
# print(iris.feature_names) It will Give Only the Feature Names
```

```python
df=pd.DataFrame({
    'x':iris.data[:,0],
    'y':iris.data[:,1],
    'cluster':iris.target
})
```

```python
centroids ={}
for i in range(3):
    result_list=[]
    result_list.append(df.loc[df['cluster'] == i]['x'].mean())
    result_list.append(df.loc[df['cluster'] == i]['y'].mean())
    centroids[i]=result_list
    
```

```python
# Calculating distance and Updating dataFrame
```

```python
def assignment(df,centroids):
    for i in centroids.keys():
        #sqrt((x1-x2)^2+(y1-y2)^2)
        df['distance_from_{}'.format(i)] = (
            np.sqrt(
                (df['x']-centroids[i][0])**2 + (df['y']-centroids[i][1])**2
            
            )
          
        )
    centroid_distance_cols=['distance_from_{}'.format(i) for i in centroids.keys()]
    df['closest'] = df.loc[:, centroid_distance_cols].idxmin(axis=1)
    df['closest'] = df['closest'].map(lambda x:int(x.lstrip('distance_from_')))
    df['color']   = df['closest'].map(lambda x: colmap[x])
    return df
```

```python
df = assignment(df, centroids)
```
